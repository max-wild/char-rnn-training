```python
import matplotlib.pyplot as plt
import json
```

```python
# Load the list from the JSON file
with open('sherlock_3.json', 'r') as file:
    sherlock_3 = json.load(file)

with open('sherlock_4.json', 'r') as file:
    sherlock_4 = json.load(file)
```

```python
plt.plot(sherlock_4['loss'], color='turquoise', label='Model 1')
plt.plot(sherlock_3['loss'], color='red', label='Model 2')

# plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.grid(linestyle='--', linewidth=1, alpha=0.5)
plt.show()
```
